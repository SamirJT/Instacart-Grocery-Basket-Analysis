### Line Charts

```python
# dataset too large, cunduct sampling function to plot line chart based on smaller dataframe that accurately represents data
```

```python
# Assign random seed number 
np.random.seed(4)
```

```python
# create a list holding true/false values to the tes np.random.rant() <= 0.7
dev = np.random.rand(len(ords_prods_merged)) <= 0.7
```

```python
# Store 70% of the sample into the dataframe big
big = ords_prods_merged[dev]
```

```python
# Store 30% of the sample into the dataframe small
small = ords_prods_merged[~dev]
```

```python
# check the big and small database
len(ords_prods_merged)
```

```python
len(big) + len(small)
```

```python
# adjusting sample to only columns necessary for the chart
df_2 = small[['orders_day_of_week','prices']]
```

```python
# create line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```
